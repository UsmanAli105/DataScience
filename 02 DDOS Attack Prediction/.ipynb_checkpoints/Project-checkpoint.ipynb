```python
cat_cols = list(n for n in df.columns if df[n].dtype == "object")
cat_cols
```

```python
size_of_df = len(df)
size_of_df
```

```python
df_altered = df.drop(['src', 'dst'], axis=1).copy()
```

```python
df_altered
```

```python
mapping = {}
for i, j in enumerate(list(df.Protocol.unique())):
    mapping[j] = i
mapping
```

```python
df_altered.Protocol = df_altered.Protocol.map(mapping)
```

```python
df_altered
```

```python
cor = df_altered.corr()
plt.figure(figsize=(20, 20))
plt.label = "Correlation of all the features"
sns.heatmap(cor, annot=True)
```

```python
df_altered.columns = df_altered.columns.astype(str)
```

```python
df_altered.columns
```

```python
for i in df_altered.columns:
    plt.figure(figsize=(8, 6))
    plt.hist(df_altered[i], edgecolor='black')
    plt.xlabel = str(i)  
    plt.title = f'Histogram of {str(i)}'
    # plt.savefig("./imgs/"+i+"_histogram.png")  
    plt.show()
```

```python
for i in df_altered.columns[0:-1]:
    plt.scatter(x=df_altered[i], y=df_altered['label'])
    plt.label=i
#     plt.savefig("./imgs/"+i+"_histogram.png")
    plt.show()
```
